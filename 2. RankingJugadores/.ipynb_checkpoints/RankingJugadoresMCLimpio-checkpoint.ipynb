```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../Wyscout/jugadoresLimpioConPosiciones.xlsx")
```

```python
df.head()
```

```python
df2 = df.copy()
```

```python
#Filtramos por posicion
df2 = df2[(df2.Posicion == "MC") & (df2.MinutosJugados >= 900)]
df2.shape
```

```python
print(df2.columns.to_list())
```

```python
#Columnas que vamos a ponderar
subdf2 = pd.DataFrame(df2,columns=['Jugador','%DuelosGanados','%PrecisionPases'
                            ,'PasesProfundidad/90','PasesProgresivos/90','JugadasClaves/90','xA/90'
                            ,'Regates/90','%DuelosAereosGanados','Interceptaciones/90'
                            ,'%PrecisionPasesLargos','SegundasAsistencias/90','TercerasAsistencias/90']) #Varias columnas
subdf2.head()
```

```python
subdf2.describe()
```

```python
#Añadimos las columnas calculadas que queramos
#subdf2['xG/RemateContra'] = df2.apply(lambda row: row['xGEnContra']/row['RematesContra'], axis=1).round(2)
```

```python
#Normalizamos cada columna por el mayor y damos el valor de la ponderizacion
subdf2['%DuelosGanados'] = ((subdf2['%DuelosGanados']/subdf2['%DuelosGanados'].max())*2).round(2)
subdf2['%PrecisionPases'] = ((subdf2['%PrecisionPases']/subdf2['%PrecisionPases'].max())*2.5).round(2)
subdf2['PasesProfundidad/90'] = ((subdf2['PasesProfundidad/90']/subdf2['PasesProfundidad/90'].max())*2).round(2)
subdf2['PasesProgresivos/90'] = ((subdf2['PasesProgresivos/90']/subdf2['PasesProgresivos/90'].max())*2).round(2)
subdf2['JugadasClaves/90'] = ((subdf2['JugadasClaves/90']/subdf2['JugadasClaves/90'].max())*2).round(2)
subdf2['xA/90'] = ((subdf2['xA/90']/subdf2['xA/90'].max())*2).round(2)
subdf2['Regates/90'] = ((subdf2['Regates/90']/subdf2['Regates/90'].max())*1.3).round(2)
subdf2['%DuelosAereosGanados'] = ((subdf2['%DuelosAereosGanados']/subdf2['%DuelosAereosGanados'].max())*1.3).round(2)
subdf2['Interceptaciones/90'] = ((subdf2['Interceptaciones/90']/subdf2['Interceptaciones/90'].max())*1.5).round(2)
subdf2['%PrecisionPasesLargos'] = ((subdf2['%PrecisionPasesLargos']/subdf2['%PrecisionPasesLargos'].max())*1.5).round(2)
subdf2['SegundasAsistencias/90'] = ((subdf2['SegundasAsistencias/90']/subdf2['SegundasAsistencias/90'].max())*1.5).round(2)
subdf2['TercerasAsistencias/90'] = ((subdf2['TercerasAsistencias/90']/subdf2['TercerasAsistencias/90'].max())*1).round(2)
```

```python
#Sumamos las columnas desde la ultima posicion hasta la que queramos para sacar la puntuacion
subdf2['PuntuacionTotal']= subdf2.iloc[:, -12:-1].sum(axis=1)
subdf2 = subdf2.sort_values(by='PuntuacionTotal', ascending=False)
subdf2.head()
```

```python
#Quitamos las columnas del dataframe original para añadir las del dataframe con la puntuacion
df2SinColumnas = df2.drop(['%DuelosGanados','%PrecisionPases'
                            ,'PasesProfundidad/90','PasesProgresivos/90','JugadasClaves/90','xA/90'
                            ,'Regates/90','%DuelosAereosGanados','Interceptaciones/90'
                            ,'%PrecisionPasesLargos','SegundasAsistencias/90','TercerasAsistencias/90'], axis=1)
df2SinColumnas.head()
```

```python
#Creamos dataframe con aspectos basicos del jugador para añadir al dataframe final
dfFinal = pd.merge(subdf2,df2SinColumnas, on='Jugador', how='left')
dfFinal.head()
```

```python
dfFinal.to_csv("puntuacionMC.csv", encoding="utf-8-sig")
```
