```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../TFM/Wyscout/jugadoresLimpioConPosiciones.xlsx")
```

```python
df.head()
```

```python
#Filtramos por posicion
df = df[(df.Posicion == "POR") & (df.MinutosJugados >= 900)]
```

```python
df.shape
```

```python
#Primer dataframe con las columnas que vamos a utilizar
subdf = pd.DataFrame(df,columns=['Jugador','Equipo','Edad','VM','MinutosJugados','GolesRecibidos','GolesRecibidos/90'
    ,'RematesContra','RematesContra/90','PorteriasImbatidas','%Paradas','xGEnContra','xGEnContra/90'
    ,'GolesEvitados','GolesEvitados/90','Salidas/90','%DuelosAereosGanados','%PrecisionPasesLargos']) #Varias columnas
```

```python
subdf.head()
```

```python
#Creacion de columnas nuevas que queramos calcular a partir de otras
subdf['xG/RemateContra'] = subdf.apply(lambda row: row['xGEnContra']/row['RematesContra'], axis=1).round(2)
subdf.head()
```

```python
subdf.describe()
```

```python
subdf.columns
```

```python
#Columnas que vamos a ponderar
subdf2 = pd.DataFrame(subdf,columns=['Jugador','%Paradas','GolesEvitados'
                            ,'%DuelosAereosGanados','%PrecisionPasesLargos','xG/RemateContra']) #Varias columnas
subdf2.head()
```

```python
#Normalizamos cada columna por el mayor y damos el valor de la ponderizacion
subdf2['%Paradas'] = ((subdf2['%Paradas']/subdf2['%Paradas'].max())*1.2).round(2)
subdf2['GolesEvitados'] = ((subdf2['GolesEvitados']/subdf2['GolesEvitados'].max())*3).round(2)
subdf2['%DuelosAereosGanados'] = ((subdf2['%DuelosAereosGanados']/subdf2['%DuelosAereosGanados'].max())*1.4).round(2)
subdf2['%PrecisionPasesLargos'] = ((subdf2['%PrecisionPasesLargos']/subdf2['%PrecisionPasesLargos'].max())*1.4).round(2)
subdf2['xG/RemateContra'] = ((subdf2['xG/RemateContra']/subdf2['xG/RemateContra'].max())*2).round(2)
subdf2.head()
```

```python
subdf2['PuntuacionTotal']= subdf2.iloc[:, -5:-1].sum(axis=1)
subdf2 = subdf2.sort_values(by='PuntuacionTotal', ascending=False)
subdf2.head()
```

```python
dfSinColumnas = df.drop(['%Paradas','GolesEvitados'
                            ,'%DuelosAereosGanados','%PrecisionPasesLargos'], axis=1)
dfSinColumnas.head()
```

```python
#Creamos dataframe con aspectos basicos del jugador para añadir al dataframe final
dfFinal = pd.merge(subdf2,dfSinColumnas, on='Jugador', how='left')
dfFinal.head()
```

```python
dfFinal.to_csv("puntuacionPorteros.csv", encoding="utf-8-sig")
```
