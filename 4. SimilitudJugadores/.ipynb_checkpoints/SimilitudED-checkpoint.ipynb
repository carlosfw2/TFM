```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("../Wyscout/Jugadores2+2BFinal.xlsx")
df.head()
```

```python
df2 = df.copy()
```

```python
#Filtramos por posicion
df2 = df2[(df2.Posicion == "MCO") & (df2.MinutosJugados >= 900)]
df2.shape
```

```python
#Columnas que vamos a ponderar
subdf2 = pd.DataFrame(df2,columns=['Remates/90','CarrerasProgresion/90'
                            ,'%RegatesRealizados','JugadasClaves/90',
                            'xG/90','GolesSinPenalti/90'
                            ,'ToquesAreaPenalti/90','xA/90',
                            'SegundasAsistencias/90','AccionesAtaqueExitosas/90'
                            ,'PasesProfundidad/90','%PrecisionPasesProfundidad']) #Varias columnas
subdf2.head()
```

```python
#Añadimos las columnas calculadas que queramos
#subdf2['xG/RemateContra'] = df2.apply(lambda row: row['xGEnContra']/row['RematesContra'], axis=1).round(2)
```

```python
#Normalizamos cada columna por el mayor y damos el valor de la ponderizacion
subdf2['Remates/90'] = ((subdf2['Remates/90']/subdf2['Remates/90'].max())*1.3).round(2)
subdf2['CarrerasProgresion/90'] = ((subdf2['CarrerasProgresion/90']/subdf2['CarrerasProgresion/90'].max())*1.5).round(2)
subdf2['%RegatesRealizados'] = ((subdf2['%RegatesRealizados']/subdf2['%RegatesRealizados'].max())*1.7).round(2)
subdf2['JugadasClaves/90'] = ((subdf2['JugadasClaves/90']/subdf2['JugadasClaves/90'].max())*2).round(2)
subdf2['xG/90'] = ((subdf2['xG/90']/subdf2['xG/90'].max())*1.7).round(2)
subdf2['GolesSinPenalti/90'] = ((subdf2['GolesSinPenalti/90']/subdf2['GolesSinPenalti/90'].max())*2).round(2)
subdf2['ToquesAreaPenalti/90'] = ((subdf2['ToquesAreaPenalti/90']/subdf2['ToquesAreaPenalti/90'].max())*1.5).round(2)
subdf2['xA/90'] = ((subdf2['xA/90']/subdf2['xA/90'].max())*3).round(2)
subdf2['SegundasAsistencias/90'] = ((subdf2['SegundasAsistencias/90']/subdf2['SegundasAsistencias/90'].max())*2).round(2)
subdf2['AccionesAtaqueExitosas/90'] = ((subdf2['AccionesAtaqueExitosas/90']/subdf2['AccionesAtaqueExitosas/90'].max())*2).round(2)
subdf2['PasesProfundidad/90'] = ((subdf2['PasesProfundidad/90']/subdf2['PasesProfundidad/90'].max())*2).round(2)
subdf2['%PrecisionPasesProfundidad'] = ((subdf2['%PrecisionPasesProfundidad']/subdf2['%PrecisionPasesProfundidad'].max())*1.5).round(2)
```

```python
#Añadimos la columna con el nombre del equipo y la colocamos en primer lugar
subdf2['Jugador'] = df2['Jugador']
first_cols = ['Jugador']
last_cols = [col for col in subdf2.columns if col not in first_cols]

subdf2 = subdf2[first_cols+last_cols]
subdf2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
subdf2.shape
```

```python
#Hacemos el PCA con el numero de componentes
def filtradoDataframe(df):
    X, y = df.iloc[:,1:12].values, df.iloc[:,0].values
    sc = StandardScaler()
    X_std = sc.fit_transform(X)
    
    pca = PCA(n_components = 8)
    pca.fit(X_std)
    X_pca = pca.transform(X_std)
    
    exp1 = pca.explained_variance_ratio_
    print('Varianza 4 componentes',sum(exp1[0:4]))
    print('Varianza 4 componentes',sum(exp1[0:8]))
    print(df.shape)
    
    df_pca_resultado = pd.DataFrame(data = X_pca, columns = ["PC1","PC2","PC3","PC4","PC5","PC6","PC7","PC8"
                                                    ], index=y)
    #df_pca_resultado = df_pca_resultado.rename_axis('Equipo')
    return df_pca_resultado.sort_values(by=['PC1'])
```

```python
df_pca_resultado = filtradoDataframe(subdf2)
df_pca_resultado.head()
```

```python
#Matriz de correlacion METODO PEARSON
df_corr = df_pca_resultado.T.corr(method='pearson')
df_corr.head()
```

```python
#Metodo1: Sin Dendogram
df_corr2 = df_corr.filter(like='Andrés Martín', axis=0)
df_corr2 =df_corr2.T
df_corr2 = df_corr2*100
```

```python
df_corr2.sort_values(by=['Andrés Martín'],ascending=False).head(11).iloc[1:]
```

```python
#Metodo2: Con Dendogram WARD
unique_index = pd.Index(list(df_corr))
unique_index.get_loc('João Carvalho')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(df_corr,"ward")
```

```python
plt.figure(figsize=(40, 15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```
