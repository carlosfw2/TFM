```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("../Wyscout/Jugadores2+2BFinal.xlsx")
df.head()
```

```python
df2 = df.copy()
```

```python
#Filtramos por posicion
df2 = df2[(df2.Posicion == "ED") & (df2.MinutosJugados >= 900)]
df2.shape
```

```python
#Columnas que vamos a ponderar
subdf2 = pd.DataFrame(df2,columns=['xG/90','Remates/90','Remates'
                            ,'xA/90','SegundasAsistencias/90',
                            '%PrecisionCentros','Desmarques/90'
                            ,'%RegatesRealizados','%DuelosOfensivosGanados'
                            ,'AccionesAtaqueExitosas/90','Regates/90'
                            ,'PasesAreaPenalti/90'
                            ,'CarrerasProgresion/90','%PrecisionPasesUltimoTercio'
                            ,'GolesSinPenalti/90','xG']) #Varias columnas
subdf2.head()
```

```python
#Añadimos las columnas calculadas que queramos
subdf2['xG/Remate'] = df2.apply(lambda row: row['xG']/row['Remates'], axis=1).round(2)
subdf2 = subdf2.drop(['xG','Remates','Remates/90'], axis=1)
```

```python
subdf2['xG/Remate'] = ((subdf2['xG/Remate']/subdf2['xG/Remate'].max())*3).round(2)
subdf2['xG/90'] = ((subdf2['xG/90']/subdf2['xG/90'].max())*3).round(2)
subdf2['xA/90'] = ((subdf2['xA/90']/subdf2['xA/90'].max())*3).round(2)
subdf2['SegundasAsistencias/90'] = ((subdf2['SegundasAsistencias/90']/subdf2['SegundasAsistencias/90'].max())*1.3).round(2)
subdf2['%PrecisionCentros'] = ((subdf2['%PrecisionCentros']/subdf2['%PrecisionCentros'].max())*1.5).round(2)
subdf2['Desmarques/90'] = ((subdf2['Desmarques/90']/subdf2['Desmarques/90'].max())*1.5).round(2)
subdf2['%RegatesRealizados'] = ((subdf2['%RegatesRealizados']/subdf2['%RegatesRealizados'].max())*2).round(2)
subdf2['%DuelosOfensivosGanados'] = ((subdf2['%DuelosOfensivosGanados']/subdf2['%DuelosOfensivosGanados'].max())*1.5).round(2)
subdf2['AccionesAtaqueExitosas/90'] = ((subdf2['AccionesAtaqueExitosas/90']/subdf2['AccionesAtaqueExitosas/90'].max())*2).round(2)
subdf2['Regates/90'] = ((subdf2['Regates/90']/subdf2['Regates/90'].max())*1.8).round(2)
subdf2['PasesAreaPenalti/90'] = ((subdf2['PasesAreaPenalti/90']/subdf2['PasesAreaPenalti/90'].max())*1.5).round(2)
subdf2['CarrerasProgresion/90'] = ((subdf2['CarrerasProgresion/90']/subdf2['CarrerasProgresion/90'].max())*1.5).round(2)
subdf2['%PrecisionPasesUltimoTercio'] = ((subdf2['%PrecisionPasesUltimoTercio']/subdf2['%PrecisionPasesUltimoTercio'].max())*1.5).round(2)
subdf2['GolesSinPenalti/90'] = ((subdf2['GolesSinPenalti/90']/subdf2['GolesSinPenalti/90'].max())*2).round(2)

subdf2.head()
```

```python
#Añadimos la columna con el nombre del equipo y la colocamos en primer lugar
subdf2['Jugador'] = df2['Jugador']
first_cols = ['Jugador']
last_cols = [col for col in subdf2.columns if col not in first_cols]

subdf2 = subdf2[first_cols+last_cols]
subdf2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
subdf2.shape
```

```python
#Hacemos el PCA con el numero de componentes
def filtradoDataframe(df):
    X, y = df.iloc[:,1:14].values, df.iloc[:,0].values
    sc = StandardScaler()
    X_std = sc.fit_transform(X)
    
    pca = PCA(n_components = 12)
    pca.fit(X_std)
    X_pca = pca.transform(X_std)
    
    exp1 = pca.explained_variance_ratio_
    print('Varianza 4 componentes',sum(exp1[0:4]))
    print('Varianza 4 componentes',sum(exp1[0:8]))
    print('Varianza 4 componentes',sum(exp1[0:12]))
    print(df.shape)
    
    df_pca_resultado = pd.DataFrame(data = X_pca, columns = ["PC1","PC2","PC3","PC4","PC5","PC6","PC7","PC8",
                                                   "PC9","PC10","PC11","PC12"], index=y)
    #df_pca_resultado = df_pca_resultado.rename_axis('Equipo')
    return df_pca_resultado.sort_values(by=['PC1'])
```

```python
df_pca_resultado = filtradoDataframe(subdf2)
df_pca_resultado.head()
```

```python
#Matriz de correlacion METODO PEARSON
df_corr = df_pca_resultado.T.corr(method='pearson')
df_corr.head()
```

```python
#Metodo1: Sin Dendogram
df_corr2 = df_corr.filter(like='José Corpas', axis=0)
df_corr2 =df_corr2.T
df_corr2 = df_corr2*100
```

```python
df_corr2.sort_values(by=['José Corpas'],ascending=False).head(11).iloc[1:]
```

```python
#Metodo2: Con Dendogram WARD
unique_index = pd.Index(list(df_corr))
unique_index.get_loc('José Corpas')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(df_corr,"ward")
```

```python
plt.figure(figsize=(40, 15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```
